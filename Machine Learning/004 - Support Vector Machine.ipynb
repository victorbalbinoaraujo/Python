```python
'''
SVM

Separa um grupo em um plano em dois ou mais de acordo
com dois pontos de grupos diferentes equidistantes a
linha traçada de separação.

A classificação ocorre quando colocamos um ponto
qualquer no plano e o modelo tenta predizer
a qual grupo pertence.

'''
```

```python
'''
A área de cada grupo precisa ser a maior possível,
dessa forma o modelo não cometa erros ao predizer.

No exemplo abaixo, o modelo prediz que o ponto preto pertence
ao grupo vermelho, pois a linha traçada ocupa a mínima
área possível, quando na realidade o ponto preto pertence
ao grupo verde.

'''
```

```python
'''
Se a linha estivesse na diagonal, por exemplo, o modelo teria
uma acurácia maior.

No exemplo abaixo, o modelo acerta ao predizer ao qual
grupo o ponto preto pertence.

'''
```

```python
'''
No mundo real, no entanto a dispersão dos dados não é tão óbvia, dados pertencentes 
a grupos distintos frequentemente estão próximos, o que atrapalha a classificação.

Para resolver este problema, é utilizado kernel, que consiste em colocar esses dados
em um plano tridimensional, facilitando o processo da separação.
Nem sempre funciona.

'''
```

```python
'''
Soft Margin

Para melhor acurácia do modelo podemos aplicar um soft margin,
função que possibilita que alguns pontos fora da linha possam 
pertencer ao seu grupo.

'''
```

```python
import sklearn
from sklearn import datasets
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import train_test_split as tts
```

```python
cancer = datasets.load_breast_cancer()

print(cancer.feature_names)
print(cancer.target_names)
```

```python
X = cancer.data
y = cancer.target

x_train, x_test, y_train, y_test = tts(X, y, test_size=0.1)
```

```python
classifier = svm.SVC(kernel='linear') 
# kernel='linear' Aplicando o espaço tridimensional para melhorar a acurácia

classifier.fit(x_train, y_train)
pred = classifier.predict(x_test)

acc = metrics.accuracy_score(y_test, pred)
acc
```
