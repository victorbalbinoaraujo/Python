```python
# Imports 
import pandas as pd

from sklearn.model_selection import train_test_split as tts
from sklearn.ensemble import ExtraTreesClassifier
```

```python
# Dataset available at ttps://www.kaggle.com/dell4010/wine-dataset

dataset = pd.read_csv('../../../CSV Files/wine_dataset.csv')
dataset
```

## Criar um modelo de ML capaz de identificar se um vinho x é branco ou vermelho baseado nas características presentes no Dataset.

### RED 1
### WHITE 0

```python
dataset['style'] = dataset['style'].replace('red', 1)
dataset['style'] = dataset['style'].replace('white', 0)
```

```python
# Separando os dados da resposta.

y = dataset['style']
x = dataset.drop(['style'], axis=1)
```

```python
print(f"""
      X: {x}
      Y: {y}
      """)
```

```python
# Criando conjuntos de treino e teste.

x_treino, x_teste, y_treino, y_teste = tts(x, y, test_size=0.3)
```

```python
print(f"""
      X_Treino: {x_treino}
      X_Teste: {x_teste}
      Y_Treino: {y_treino}
      Y_Teste: {y_teste}
    """)
```

```python
model = ExtraTreesClassifier()

model.fit(x_treino, y_treino)
score = model.score(x_teste, y_teste)

print(f"A acurácia do modelo foi de: {round(score*100)}%")
```
