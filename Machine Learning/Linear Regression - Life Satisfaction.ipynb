```python
# Imports

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
```

```python
# Load data.

# Reading CSV file.
df = pd.read_csv('../../CSV Files/gdp-life-satisfaction.csv', sep=';')

# Removing NaN values.
df = df.dropna()

# Selecting a Year.
df_2017 = df.loc[(df['Year'] == 2017)]

# Comparing.
countries = set(df['Entity'])
countries_2017 = df_2017['Entity']

# Difference between df's.
len(countries) - len(countries_2017) # 16 countries.

df = df_2017
df
```

```python
# Renaming columns. 
df = df.rename(columns={
    'Entity' : 'country',
    'GDP per capita (constant 2011 international $)' : 'gdp',
    ' Life satisfaction (country average': 'satisfaction',
}, inplace=False)

# Removing useless columns.
df = df.drop(['Code', 'Year'], axis=1)
```

```python
df
```

```python
# Visualizing
df.plot(kind='scatter', x='gdp', y='satisfaction')
plt.show()
```

```python
# Selecting a linear model.
model = sklearn.linear_model.LinearRegression()

X = np.c_[df.gdp]
y = np.c_[df.satisfaction]

# Training.
model.fit(X, y)
```

```python
# Making predictions.

X_input = float(input())

X_new = [[X_input]]
print(model.predict(X_new))
```
