```python
"""
KNN

Ao inserir um novo dado em um espaço onde 
já estão inseridos outros grupos de dados próximos
prever a qual grupo este novo dado pertence.

Seleciona o grupo de classes mais próxima e 
atribui o novo dado a este grupo.

K determina a quantidade de dados a serem analisados para
determinar o novo dado. O grupo que conter o maior número de dados
próximo ao novo dado determinará a classe.
Por critérios de desempate K deve ser SEMPRE um número ímpar.

O cálculo para classificação do novo ponto é feito, por padrão,
utilizando a distância euclidiana.

"""
```

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier as knc
from sklearn import linear_model, preprocessing
from sklearn.model_selection import train_test_split as tts
```

```python
df = pd.read_csv('../CSV Files/car.data')
df
```

```python
'''
Para este dataset em específico, é necessário converter os valores não-numéricos em numéricos (0, 1, 2, ...).
Exemplo: Safety(low = 0, med = 1, high = 2).
Para isto, o preprocessing é utilizado.

'''
prep = preprocessing.LabelEncoder()

buying = prep.fit_transform(list(df['buying']))
maint = prep.fit_transform(list(df['maint']))
door = prep.fit_transform(list(df['door']))
persons = prep.fit_transform(list(df['persons']))
lug_boot = prep.fit_transform(list(df['lug_boot']))
safety = prep.fit_transform(list(df['safety']))
cls = prep.fit_transform(list(df['class']))
```

```python
X = list(zip(buying, maint, door, persons, lug_boot, safety))
y = list(cls)

x_train, x_test, y_train, y_test = tts(X, y, test_size=0.1)
```

```python
model = knc(n_neighbors=9) # Parâmetro: quantidade de vizinhos. 
# A acúracia pode aumentar conforme a quantidade de vizinhos aumenta.
# Até chegar um limite, neste caso 9, a partir do 11 decresce.

model.fit(x_train, y_train)
acc = model.score(x_test, y_test)
print(acc)
```

```python
pred = model.predict(x_test)
acertos = 0
for i in range(len(pred)):
    # print(f'Valor previsto: {round(pred[i])} Valor correto: {y_test[i]}') Comentado para a visualização ser melhor.
    # Distância:
    n = model.kneighbors([x_test[i]], 9, True)
    print(f'N : {n}')
    if round(pred[i]) == y_test[i]:
        acertos += 1
print(f'{acertos} acertos de {len(pred)}')
```
