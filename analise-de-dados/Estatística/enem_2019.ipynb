```python
df = df.rename(columns={
    'NU_NOTA_REDACAO' : 'NOTA_REDACAO',
    'NU_NOTA_CN' : 'NOTA_CN',
    'NU_NOTA_CH' : 'NOTA_CH',
    'NU_NOTA_LC' : 'NOTA_LC',
    'NU_NOTA_MT' : 'NOTA_MT',
    'NU_NOTA_COMP1' : 'COMP1',
    'NU_NOTA_COMP2' : 'COMP2',
    'NU_NOTA_COMP3' : 'COMP3',
    'NU_NOTA_COMP4' : 'COMP4',
    'NU_NOTA_COMP5' : 'COMP5',
    'NU_IDADE' : 'IDADE',
    'TP_SEXO' : 'SEXO',
    'TP_COR_RACA' : 'RACA',
    'Q025' : 'INTERNET',
    'TP_ESCOLA' : 'ESCOLA'
})
```

```python
df
```

```python
df['RACA'] = df['RACA'].replace({
    0:'nao_declarado',
    1: 'branca',
    2: 'preta',
    3: 'parda',
    4: 'amarela',
    5: 'indigena'
})
```

```python
df['TP_LINGUA'] = df['TP_LINGUA'].replace({
    0: 'Inglês',
    1: 'Espanhol'
})
```

```python
df['ESCOLA'] = df['ESCOLA'].replace({
    1: 'não_respondeu',
    2: 'pública',
    3: 'privada',
    4: 'exterior'
})
```

```python
df['INTERNET'] = df['INTERNET'].replace({
    'A': 'Não',
    'B': 'Sim'
})
```
